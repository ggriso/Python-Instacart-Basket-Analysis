## 05. Create line charts

```python
# Create a 70/30 sample

np.random.seed(4)
```

```python
# Create a list holding True/False values to the test np.random.rand() <= 0.7

dev = np.random.rand(len(df_final)) <= 0.7
```

```python
dev
```

```python
# Store 70% of the sample in the dataframe big

big = df_final[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = df_final[~dev]
```

```python
# Check the output

len(df_final)
```

```python
len(big)+len(small)
```

```python
# Reduce the samples to only those columns necessary for the chart

df = small[['orders_day_of_week','prices']]
```

```python
# Create a line chart

line = sns.lineplot(data = df, x = 'orders_day_of_week', y = 'prices')
```

```python
# Export the chart

line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_chart_prices.png'))
```
